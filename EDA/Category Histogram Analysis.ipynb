# Category Histogram Analysis
---

## Import

```python
import numpy as np
import os
import seaborn
import json
import multiprocessing as mp
from tqdm import tqdm
import math
import pandas as pd
```

## Config

```python
CONFIG = {
    "path_data":"../../../dataset/",
    "info_data_train":"train.json",  # images, annotations
    "class":['General trash', 'Paper', 'Paper pack', 'Metal', 'Glass', 'Plastic', 'Styrofoam', 'Plastic bag', 'Battery', 'Clothing']
}
```

## Load Json

```python
with open(os.path.join(CONFIG['path_data'], CONFIG['info_data_train'])) as js:
    json_train = json.load(js)
info_train = json_train['images']
annotations_train = json_train['annotations']
```

## Make DataFrame

```python
df = pd.DataFrame(annotations_train)
df['Category'] = [CONFIG['class'][i] for i in df['category_id'].values]
```

## Render Plot

```python
seaborn.set(rc={'figure.figsize':(10,7)})
plot = seaborn.histplot(df, x='Category', hue='Category', kde=True)
plot.set_xticklabels(plot.get_xticklabels(), rotation=30)
```
